```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
file_path = r'C:\Users\DELL\Downloads\archive\extracted_data\online_shoppers_intention.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
print("First 5 rows:")
print(df.head())
```

```python
print("/nMissing values:")
print(df.isnull().sum())
```

```python
print("/nDuplicates rows:",
df.duplicated().sum())
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
categorical_cols = ['Month', 'VisitorType', 'Weekend']
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
df['Revenue'] = df['Revenue'].astype(int)
```

```python
sns.countplot(x='Revenue', data=df)
plt.title('Purchasing Intention Distribution')
plt.show
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True,
cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.dtypes
```

```python
df['Month']=df['Month'].astype('category')
df['VisitorType']=df['VisitorType'].astype('category')
df['Weekend']=df['Weekend'].astype('bool')
df['Revenue']=df['Revenue'].astype('bool')
```

```python
df.dtypes
```

```python
cleaned_file = r'C:\Users\DELL\Downloads\archive\extracted_data\cleaned_online_shoppers_intention.csv'
df.to_csv(cleaned_file, index=False)
```
